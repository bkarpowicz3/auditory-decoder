```python
def create_sine(f, shift):
    Fs=1000
    duration = 2 #seconds 
    samples = duration * Fs
    t = np.arange(start = 0, stop = duration, step = duration / samples)
    y = np.sin(2 * np.pi * f * t + shift)
    return t, y
```

```python
from scipy.spatial.distance import euclidean
from fastdtw import fastdtw

def mydist(x,y):
    distance, path = fastdtw(x, y, dist=euclidean)
    return distance 
```

## Neighbors 

```python
#training is list of files to train on 
#testInstance is a single file to compute k nearest neighbors on
#returns the categorical votes for that item

import operator
def get_neighbors(training, trainingLabels, testInstance, k):
    test = scipy.io.loadmat(testInstance)['block']
    
    distances = []
    for x in range(len(training)):
        train = scipy.io.loadmat(training[x])['block']
        d = mydist(test, train)
        distances.append((trainingLabels[x], d))
        
    neighbors = []
    distances.sort(key=operator.itemgetter(1))
    for i in range(k):
        neighbors.append(distances[x][0])
    return neighbors
```

## Classify

```python
def get_response(neighbors):
    classVotes = {}
    for x in range(len(neighbors)):
        response = neighbors[x][-1]
        if response in classVotes:
            classVotes[response] += 1
        else: 
            classVotes[response] = 1
    sortedVotes = sorted(classVotes.items(), key = operator.itemgetter(1), reverse = True)
    return sortedVotes[0][0]
```

```python
def evaluate(testLabels, predictions):
    correct = 0
    for x in range(len(testLabels)):
        if testLabels[x] == predictions[x]: 
            correct += 1
    return correct/float(len(testLabels)) * 100.0
```

```python
all_accuracy = []
for train_index, test_index in kf: 
    X_train = np.array(data)[train_index]
    X_test = np.array(data)[test_index]
    y_train = np.array(labels)[train_index]
    y_test = np.array(labels)[test_index]
    
    predictions = []
    k = 5
    for x in range(len(X_test)):
        print('Classifying ' + str(x+1) + ' of ' + str(len(X_test)))
        neighbors = get_neighbors(X_train, y_train, X_test[x], k)
        result = get_response(neighbors)
        predictions.append(result)
        print('Predicted: ' + result + ', Actual: ' + y_test[x])
    accuracy = evaluate(y_test, predictions)
    all_accuracy.append(accuracy)
    print('Accuracy: ' + str(accuracy))

total_acc = np.mean(all_accuracy)
print('Total Accuracy: ' + str(total_acc))
```
